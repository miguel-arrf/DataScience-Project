```python
import altair as alt

def get_variable_types(df) -> dict:
    variable_types: dict = {
        'Numeric': [],
        'Binary': [],
        'Date': [],
        'Symbolic': []
    }
    for c in df.columns:
        uniques = df[c].dropna(inplace=False).unique()
        if len(uniques) == 2:
            variable_types['Binary'].append(c)
            df[c].astype('bool')
        elif df[c].dtype == 'datetime64':
            variable_types['Date'].append(c)
        elif df[c].dtype == 'int':
            variable_types['Numeric'].append(c)
        elif df[c].dtype == 'float':
            variable_types['Numeric'].append(c)
        else:
            df[c].astype('category')
            variable_types['Symbolic'].append(c)

    return variable_types

variable_types = get_variable_types(dataframe)
print(variable_types)
counts = {}
for tp in variable_types.keys():
    counts[tp] = len(variable_types[tp])

df1 = pd.DataFrame(list(zip(list(counts.keys()), list(counts.values()))),  columns =['Name', 'val'])
    
print(df1)
bars = alt.Chart(df1, title="Number of variables per type").mark_bar(fill="#ff7675",
    cornerRadiusTopLeft=0,
    cornerRadiusTopRight=0,
                                                                   size=15
                                 
).encode(
    x=alt.X('Name:O',axis=alt.Axis(title='', titleColor='#fdcb6e', labelAngle=0)),
    y = alt.Y('val:Q',
          axis=alt.Axis(title='', titleColor='#fdcb6e', labelAngle=0,),
          ),
)

text = bars.mark_text(
    align='center',
    baseline='middle',
    dx=0,# Nudges text to right so it doesn't appear on top of the bar
    dy= -3
).encode(
    text='val:Q'
)

(bars + text).configure_axis(
    grid=False
).properties(width=250,
    height=150)
```
